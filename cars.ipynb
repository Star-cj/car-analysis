```python
#Grouping of the make based on the HP and MPG. focusing on decending of HP. the reset_index() will make the index count to appear at the left
hp_mpg=car.groupby(by="Make").max()[["HP","MPG"]].sort_values(by=['HP'],ascending=False).reset_index()
hp_mpg.head(8)
```

```python
#Growth trend for HP and MPG (line graph)
from matplotlib.pyplot import figure


fig=plt.figure(figsize=(12,6))
ax=sns.lineplot(data=hp_mpg)
```

```python
#To create trends of bars for HP to Make
hp_mpg1=car.groupby(by="Make").max()[["HP","MPG"]].sort_values(by=['HP'],ascending=False).reset_index()
fig=plt.figure(figsize=(20,6))
ax=sns.barplot(data=hp_mpg1.iloc[:12], y="HP",x="Make",linewidth=2,edgecolor="black" ) #[:12] how many bars to be included
```

```python
#pie chart of cars

cylinder=car['Cylinders'].sum()
cityM=car['CityMpg'].sum()
mpg=car['MPG'].sum()

px.pie(names=['MPG','CityMPG','Cylinder'],values=[cylinder,cityM,mpg],title="A pie chart of car property")
# ,autopct='%0.1f%%', explode=(0,0.2,0)
```

```python
# Most Make car from the data
sort_car=car.groupby('Make')['MSRP'].sum().to_frame('MSRP')   #to group all the Make of the same name and sum the MSRP of that make
sort_car=sort_car.sort_values('MSRP', ascending=False)[:8]  #to arrange in decending order and [:8] is for first 8 and [-5:] means the last 5
sort_car
```

```python
# The plot of sorted cars by MSRP 
from numpy import size


plt.figure(figsize=(12,6))
plt.title('The top 7 MSRP of Cars', size=30)
sns.barplot(data=sort_car, y=sort_car.MSRP,x=sort_car.index,linewidth=2,edgecolor="black")
plt.xlabel('Car Makes', size=20)
plt.ylabel('De MSRP', size=20)
plt.show()
```

```python
# To show the last 5 cars based on msrp 
last_msrp_car=car.groupby('Make')['MSRP'].sum().to_frame('MSRP')  
last_msrp_car=last_msrp_car.sort_values('MSRP', ascending=False)[-5:]  #[-5] the last 5 cars

plt.figure(figsize=(10,6))
plt.title('The last 5 Cars based on MSRP', size=30)
sns.barplot(data=last_msrp_car, y=last_msrp_car.MSRP,x=last_msrp_car.index,linewidth=2,edgecolor="black")
plt.xlabel('Car Makes', size=20)
plt.ylabel('De MSRP', size=20)
plt.show()
```
